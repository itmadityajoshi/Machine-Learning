1. Outlier detection using Z score
    68% of data points will be in the region of 1 standard deviation.
    95% of data points will be in the region of 2 standard deviation.
    99.7% of data points will be in the region of 3 standard deviation.

 formula  z = x- mean / standard deviation

 
Reference to image:
 When the population mean and the population standard deviation are unknown, the standard score may be calculated using the sample mean (x) and sample standard deviation (s) as estimates of the population values.




```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stat
```

```python
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]
print(dataset)
```

```python

sns.histplot(data=dataset, color='blue', kde= True)
```

above distribution is not normally distributed

```python
mean = np.mean(dataset)
# print(mean)

deviation = stat.stdev(dataset)
# print(deviation)

# print(len(dataset))
out = mean + 3*deviation
print(out)
outlier = []
for i in dataset:
    x = mean + 3*deviation
    if i > x:
        outlier.append(i) 

print(outlier)
```

```python
q3 = np.percentile(dataset,75)
print(q3)


q1 = np.percentile(dataset,25)
print(q1)

IQR = q3 - q1 
print(IQR)

lower_fence = q1 - 1.5 * IQR
upper_fence = q3 + 1.5 * IQR

outlierss = []
for i in dataset:
    if i > upper_fence:
        outlierss.append(i)
    else:
        if i < lower_fence:
            outlierss.append(i)

print(outlierss)
```
